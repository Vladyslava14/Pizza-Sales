```python
import pandas as pd
```

```python
df = pd.read_csv('/content/pizza_sales.csv', encoding='utf-8')
df.head()
```

```python
df2 = df.dropna(axis=1)
df2.head()
```

# Перетворення дати у формат YYYY-MM-DD

```python
df2['order_date'] = pd.to_datetime(df2['order_date'], dayfirst=True).dt.date
```

```python
df2.head()
```

# Замінити кому на крапку в числових значеннях (float)

```python
df2['unit_price'] = df2['unit_price'].astype(str).str.replace(',', '.').astype(float)
df2['total_price'] = df2['total_price'].astype(str).str.replace(',', '.').astype(float)
df2.head()
```

# Перетворення у тип float

```python
df2['unit_price'] = df2['unit_price'].astype(float)
df2['total_price'] = df2['total_price'].astype(float)
df2.head()
```

Перевірка на пропущені значення

```python
df2.isna().sum()
```

```python
df2 = df2.drop_duplicates()
```

```python
# 3. Примусове перетворення до float
df2['unit_price'] = pd.to_numeric(df2['unit_price'], errors='coerce')
df2['total_price'] = pd.to_numeric(df2['total_price'], errors='coerce')

# 4. Видаляємо рядки з некоректними значеннями (NaN після перетворення)
df2 = df2.dropna(subset=['unit_price', 'total_price'])
```

```python
# Видаляємо рядки з порожніми значеннями
df2.dropna(how='any', inplace=True)

# Скидаємо індекси
df2.reset_index(drop=True, inplace=True)

# Зберігаємо очищений CSV
df2.to_csv('pizza_cleaned.csv', index=False, lineterminator='\n')
```
